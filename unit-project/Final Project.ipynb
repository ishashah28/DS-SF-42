```python
feature_col = ['OverallQual','YearBuilt','YrSold','TotRmsAbvGrd','GarageCars','FullBath','TotalBsmtSF']
X = df[feature_col]
y = df['SalePrice']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.40, random_state = 30)

lr = LinearRegression()
lr.fit(X_train, y_train)

#training score
train_score = lr.score(X_train, y_train)
print("R2 training score : {}".format(train_score))
test_score = lr.score(X_test, y_test)
print("R2 testing score : {}".format(test_score))
```

```python
#cross validatio
cross_val_score(LinearRegression(), X, y, cv=5, scoring = "r2").mean()
```
